```python
#Soal 1 Kalkulator 
import tkinter as tk

def Kalkulator():
    def calculate(operation):
        try:
            num1 = float(entry1.get())
            num2 = float(entry2.get())
            if operation == 'add':
                result = num1 + num2
            elif operation == 'subtract':
                result = num1 - num2
            elif operation == 'multiply':
                result = num1 * num2
            elif operation == 'divide':
                if num2 != 0:
                    result = num1 / num2
                else:
                    label_result.config(text='Error: Division by zero')
                    return
            label_result.config(text=f'Result: {result}')
        except ValueError:
            label_result.config(text='Error: Masukkan angka yang valid')

    root = tk.Tk()
    root.title('Kalkulator')

    # Label dan input untuk angka pertama
    tk.Label(root, text='Number 1:').grid(row=0, column=0, padx=5, pady=5)
    entry1 = tk.Entry(root)
    entry1.grid(row=0, column=1, padx=5, pady=5)

    # Label dan input untuk angka kedua
    tk.Label(root, text='Number 2:').grid(row=1, column=0, padx=5, pady=5)
    entry2 = tk.Entry(root)
    entry2.grid(row=1, column=1, padx=5, pady=5)

    # Frame untuk tombol operasi
    frame_buttons = tk.Frame(root)
    frame_buttons.grid(row=2, column=0, columnspan=2, pady=5)

    # Tombol operasi 
    tk.Button(frame_buttons, text='+', command=lambda: calculate('add')).grid(row=0, col--umn=0, padx=5)
    tk.Button(frame_buttons, text='-', command=lambda: calculate('subtract')).grid(row=0, column=1, padx=5)
    tk.Button(frame_buttons, text='*', command=lambda: calculate('multiply')).grid(row=0, column=2, padx=5)
    tk.Button(frame_buttons, text='/', command=lambda: calculate('divide')).grid(row=0, column=3, padx=5)

    # Label untuk menampilkan hasil
    label_result = tk.Label(root, text='Result:')
    label_result.grid(row=3, column=0, columnspan=2, pady=5)

    root.mainloop()

Kalkulator()
```

```python
#Soal 2 To DoList
import tkinter as tk
from tkinter import messagebox, filedialog

def latihan_11():
    root = tk.Tk()
    root.title('Latihan 11: To-Do List')

    tasks = []

    def tambah_tugas():
        tugas = entry_tugas.get()
        if tugas:
            tasks.append(tugas)
            update_listbox()
            entry_tugas.delete(0, tk.END)
    
    def hapus_tugas():
        selected_task = listbox_tugas.curselection()
        if selected_task:
            tasks.pop(selected_task[0])
            update_listbox()

    def update_listbox():
        listbox_tugas.delete(0, tk.END)
        for task in tasks:
            listbox_tugas.insert(tk.END, task)

    def simpan_tugas():
        file_path = filedialog.asksaveasfilename(defaultextension='.txt', 
        filetypes=[('Text files', '*.txt')])
        if file_path:
            try:
                with open(file_path, 'w') as file:
                    for task in tasks:
                        file.write(task + '\n')
                messagebox.showinfo('Success', 'Tasks saved successfully.')
            except Exception as e:
                messagebox.showerror('Error', f'Error saving tasks: {str(e)}')

    def buka_tugas():
        file_path = filedialog.askopenfilename(filetypes=[('Text files', '*.txt')])
        if file_path:
            try:
                with open(file_path, 'r') as file:
                    loaded_tasks = file.readlines()
                    tasks.clear()
                    for task in loaded_tasks:
                        tasks.append(task.strip())
                    update_listbox()
                messagebox.showinfo('Success', 'Tasks loaded successfully.')
            except Exception as e:
                messagebox.showerror('Error', f'Error loading tasks: {str(e)}')

    # GUI components
    tk.Label(root, text="Masukkan Tugas:").grid(row=0, column=0, padx=10, pady=5)
    entry_tugas = tk.Entry(root)
    entry_tugas.grid(row=0, column=1, padx=10, pady=5)

    tombol_tambah = tk.Button(root, text='Tambah Tugas', command=tambah_tugas)
    tombol_tambah.grid(row=1, column=0, columnspan=2, pady=5)

    listbox_tugas = tk.Listbox(root)
    listbox_tugas.grid(row=2, column=0, columnspan=2, padx=10, pady=5)

    tombol_hapus = tk.Button(root, text='Hapus Tugas', command=hapus_tugas)
    tombol_hapus.grid(row=3, column=0, columnspan=2, pady=5)

    tombol_simpan = tk.Button(root, text='Simpan Tugas', command=simpan_tugas)
    tombol_simpan.grid(row=4, column=0, padx=10, pady=5)

    tombol_buka = tk.Button(root, text='Buka Tugas', command=buka_tugas)
    tombol_buka.grid(row=4, column=1, padx=10, pady=5)

    root.mainloop()

latihan_11()
```

```python
#Soal 3 Buku Kontak
import tkinter as tk
from tkinter import messagebox

def latihan_12():
    root = tk.Tk()
    root.title('Latihan 12: Buku Telepon')
    kontak = []

    def tambah_kontak():
        nama = entry_nama.get().strip()
        nomor = entry_nomor.get().strip()
        
        # Validasi: Nama tidak boleh kosong dan nomor telepon harus angka
        if not nama:
            messagebox.showerror("Error", "Nama tidak boleh kosong")
        elif not nomor.isdigit():
            messagebox.showerror("Error", "Nomor telepon harus berisi angka")
        else:
            kontak.append((nama, nomor))
            update_listbox()
            entry_nama.delete(0, tk.END)
            entry_nomor.delete(0, tk.END)

    def hapus_kontak():
        selected_kontak = listbox_kontak.curselection()
        if selected_kontak:
            kontak.pop(selected_kontak[0])
            update_listbox()

    def update_listbox():
        listbox_kontak.delete(0, tk.END)
        for item in kontak:
            listbox_kontak.insert(tk.END, f'{item[0]} - {item[1]}')

    def edit_kontak():
        selected_kontak = listbox_kontak.curselection()
        if selected_kontak:
            nama_baru = entry_nama.get().strip()
            nomor_baru = entry_nomor.get().strip()

            # Validasi saat mengedit kontak
            if not nama_baru:
                messagebox.showerror("Error", "Nama tidak boleh kosong")
            elif not nomor_baru.isdigit():
                messagebox.showerror("Error", "Nomor telepon harus berisi angka")
            else:
                kontak[selected_kontak[0]] = (nama_baru, nomor_baru)
                update_listbox()
                entry_nama.delete(0, tk.END)
                entry_nomor.delete(0, tk.END)

    def load_selected_kontak(event):
        selected_kontak = listbox_kontak.curselection()
        if selected_kontak:
            nama, nomor = kontak[selected_kontak[0]]
            entry_nama.delete(0, tk.END)
            entry_nama.insert(0, nama)
            entry_nomor.delete(0, tk.END)
            entry_nomor.insert(0, nomor)

    # GUI Components
    tk.Label(root, text="Nama:").grid(row=0, column=0, padx=10, pady=5, sticky='e')
    entry_nama = tk.Entry(root, width=40)
    entry_nama.grid(row=0, column=1, padx=10, pady=5)

    tk.Label(root, text="Nomor Telepon:").grid(row=1, column=0, padx=10, pady=5, sticky='e')
    entry_nomor = tk.Entry(root, width=40)
    entry_nomor.grid(row=1, column=1, padx=10, pady=5)

    frame_tombol = tk.Frame(root)
    frame_tombol.grid(row=2, column=0, columnspan=2, pady=10)

    tombol_tambah = tk.Button(frame_tombol, text="Tambah", command=tambah_kontak)
    tombol_tambah.grid(row=0, column=0, padx=5)

    tombol_edit = tk.Button(frame_tombol, text="Edit", command=edit_kontak)
    tombol_edit.grid(row=0, column=1, padx=5)

    tombol_hapus = tk.Button(frame_tombol, text="Hapus", command=hapus_kontak)
    tombol_hapus.grid(row=0, column=2, padx=5)

    listbox_kontak = tk.Listbox(root, width=50, height=10)
    listbox_kontak.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

    # Event binding untuk mengisi field saat kontak dipilih
    listbox_kontak.bind('<<ListboxSelect>>', load_selected_kontak)

    # Mengatur proporsi baris dan kolom
    root.grid_rowconfigure(2, weight=1)
    root.grid_columnconfigure(1, weight=1)

    root.mainloop()

latihan_12()
```

```python
#Soal 4 Voting
import tkinter as tk
from tkinter import messagebox
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

def voting():
    root = tk.Tk()
    root.title("Aplikasi Voting Sederhana")

    # Inisialisasi jumlah suara untuk 3 kandidat
    suara_kandidat = {
        "Kandidat A": 0,
        "Kandidat B": 0,
        "Kandidat C": 0
    }

    def update_chart():
        # Membuat figure untuk grafik
        fig, ax = plt.subplots()

        # Menampilkan bar chart
        kandidat = list(suara_kandidat.keys())
        jumlah_suara = list(suara_kandidat.values())
        ax.bar(kandidat, jumlah_suara, color=['blue', 'green', 'red'])

        # Menampilkan angka di atas batang
        for i, v in enumerate(jumlah_suara):
            ax.text(i, v + 0.1, str(v), ha='center')

        ax.set_title("Hasil Voting")
        ax.set_ylabel("Jumlah Suara")

        # Menghapus canvas lama jika ada, lalu menambahkan yang baru
        for widget in frame_chart.winfo_children():
            widget.destroy()

        canvas = FigureCanvasTkAgg(fig, master=frame_chart)
        canvas.get_tk_widget().pack()
        canvas.draw()

    def vote_kandidat(kandidat):
        suara_kandidat[kandidat] += 1
        update_chart()

    # Membuat frame untuk tombol voting
    frame_tombol = tk.Frame(root)
    frame_tombol.pack()

    tk.Label(root, text="Pilih Kandidat:").pack(pady=10)

    # Membuat tombol untuk setiap kandidat
    for kandidat in suara_kandidat.keys():
        tk.Button(frame_tombol, text=kandidat, command=lambda k=kandidat: vote_kandidat(k)).pack(side=tk.LEFT, padx=5)

    # Membuat frame untuk menampilkan grafik
    frame_chart = tk.Frame(root)
    frame_chart.pack(pady=20)

    # Inisialisasi tampilan chart pertama kali
    update_chart()

    root.mainloop()

voting()
```

```python
#Soal 5 Booking Film
import tkinter as tk
from tkinter import messagebox

def aplikasi_penyewaan_film():
    root = tk.Tk()
    root.title("Aplikasi Penyewaan Film")

    # Daftar film (judul, genre, status (tersedia/tersewa))
    film_list = [
        {"judul": "Lembayung", "genre": "Horor", "status": "Tersedia"},
        {"judul": "Agak Laen", "genre": "Comedy", "status": "Tersedia"},
        {"judul": "Seni Memahami Kekasih", "genre": "Drama", "status": "Tersedia"},
        {"judul": "Fabricated City", "genre": "Action", "status": "Tersedia"},
        {"judul": "Kang Mak Pee Mak", "genre": "Comedy", "status": "Tersedia"}
    ]

    selected_genre = tk.StringVar(value="All")

    def update_listbox():
        listbox_film.delete(0, tk.END)
        for film in film_list:
            if selected_genre.get() == "All" or film["genre"] == selected_genre.get():
                listbox_film.insert(tk.END, f'{film["judul"]} ({film["genre"]}) - {film["status"]}')

    def sewa_film():
        selected = listbox_film.curselection()
        if selected:
            index = selected[0]
            film = film_list[index]
            if film["status"] == "Tersedia":
                film["status"] = "Tersewa"
                update_listbox()
            else:
                messagebox.showwarning("Info", f'{film["judul"]} sudah tersewa.')

    def kembalikan_film():
        selected = listbox_film.curselection()
        if selected:
            index = selected[0]
            film = film_list[index]
            if film["status"] == "Tersewa":
                film["status"] = "Tersedia"
                update_listbox()
            else:
                messagebox.showwarning("Info", f'{film["judul"]} belum disewa.')

    def tambah_film():
        judul = entry_judul.get()
        genre = entry_genre.get()
        if judul and genre:
            film_list.append({"judul": judul, "genre": genre, "status": "Tersedia"})
            update_listbox()
            entry_judul.delete(0, tk.END)
            entry_genre.delete(0, tk.END)
        else:
            messagebox.showwarning("Error", "Judul dan Genre tidak boleh kosong.")

    def hapus_film():
        selected = listbox_film.curselection()
        if selected:
            index = selected[0]
            del film_list[index]
            update_listbox()

    def filter_genre(event):
        update_listbox()

    # Dropdown untuk memilih genre
    tk.Label(root, text="Filter Berdasarkan Genre:").grid(row=0, column=0, padx=10, pady=10, sticky='w')
    genre_dropdown = tk.OptionMenu(root, selected_genre, "All", "Action", "Comedy", "Drama", command=filter_genre)
    genre_dropdown.grid(row=0, column=1, padx=10, pady=10, sticky='w')

    # Listbox untuk menampilkan daftar film
    listbox_film = tk.Listbox(root, height=10, width=50)
    listbox_film.grid(row=1, column=0, columnspan=2, padx=10, pady=10)

    # Tombol Sewa dan Kembalikan
    frame_tombol = tk.Frame(root)
    frame_tombol.grid(row=3, column=0, columnspan=2, pady=10)

    tombol_sewa = tk.Button(frame_tombol, text="Sewa", command=sewa_film)
    tombol_sewa.grid(row=0, column=0, padx=5)

    tombol_kembalikan = tk.Button(frame_tombol, text="Kembalikan", command=kembalikan_film)
    tombol_kembalikan.grid(row=0, column=1, padx=5)

   
    # Form untuk menambah film
    tk.Label(root, text="Tambah Film:").grid(row=4, column=0, columnspan=2, pady=10, sticky='w')

    tk.Label(root, text="Judul Film:").grid(row=5, column=0, padx=10, pady=5, sticky='w')
    entry_judul = tk.Entry(root)
    entry_judul.grid(row=5, column=1, padx=10, pady=5, sticky='w')

    tk.Label(root, text="Genre Film:").grid(row=6, column=0, padx=10, pady=5, sticky='w')
    entry_genre = tk.Entry(root)
    entry_genre.grid(row=6, column=1, padx=10, pady=5, sticky='w')
    
    frame_tambah = tk.Frame(root)
    frame_tambah.grid(row=7, column=0, columnspan=2, pady=10)

    tombol_tambah = tk.Button(frame_tambah, text="Tambah Film", command=tambah_film)
    tombol_tambah.grid(row=0, column=0, padx=5)

    tombol_hapus = tk.Button(frame_tambah, text="Hapus Film", command=hapus_film)
    tombol_hapus.grid(row=0, column=1, padx=5)

    # Inisialisasi tampilan awal listbox
    update_listbox()

    root.mainloop()

aplikasi_penyewaan_film()
```

```python
#Soal 6 Kalkulator BMI
import tkinter as tk
from tkinter import messagebox

def kalkulator_bmi():
    def hitung_bmi():
        try:
            berat = float(entry_berat.get())
            tinggi = float(entry_tinggi.get())
            
            if berat <= 0 or tinggi <= 0:
                messagebox.showerror("Error", "Berat dan tinggi harus lebih dari 0.")
                return

            # Menghitung BMI dengan rumus yang benar
            bmi = berat / (tinggi ** 2)
            label_hasil_bmi.config(text=f"BMI Anda: {bmi:.2f}")

            # Menentukan kategori BMI
            if bmi < 18.5:
                kategori = "Underweight"
            elif 18.5 <= bmi < 24.9:
                kategori = "Normal"
            elif 25 <= bmi < 29.9:
                kategori = "Overweight"
            else:
                kategori = "Obese"
            
            label_hasil_kategori.config(text=f"Kategori: {kategori}")
        except ValueError:
            messagebox.showerror("Error", "Masukkan angka yang valid untuk berat dan tinggi.")

    root = tk.Tk()
    root.title("Kalkulator BMI")

    # Label dan Entry untuk berat badan (kg)
    tk.Label(root, text="Berat (kg):").grid(row=0, column=0, padx=10, pady=10)
    entry_berat = tk.Entry(root)
    entry_berat.grid(row=0, column=1, padx=10, pady=10)

    # Label dan Entry untuk tinggi badan (m)
    tk.Label(root, text="Tinggi (m):").grid(row=1, column=0, padx=10, pady=10)
    entry_tinggi = tk.Entry(root)
    entry_tinggi.grid(row=1, column=1, padx=10, pady=10)

    # Tombol untuk menghitung BMI
    tombol_hitung = tk.Button(root, text="Hitung BMI", command=hitung_bmi)
    tombol_hitung.grid(row=2, column=0, columnspan=2, pady=10)

    # Label untuk hasil BMI
    label_hasil_bmi = tk.Label(root, text="BMI Anda: ")
    label_hasil_bmi.grid(row=3, column=0, columnspan=2, pady=5)

    # Label untuk kategori BMI
    label_hasil_kategori = tk.Label(root, text="Kategori: ")
    label_hasil_kategori.grid(row=4, column=0, columnspan=2, pady=5)

    root.mainloop()

kalkulator_bmi()
```

```python
#Soal 7 Konversi Mata Uang
import tkinter as tk
from tkinter import ttk

# Daftar nilai tukar tetap (contoh)
exchange_rates = {
    'IDR': {'USD': 0.000066, 'EUR': 0.000056},
    'USD': {'IDR': 15250, 'EUR': 0.85},
    'EUR': {'IDR': 17900, 'USD': 1.18}
}

def convert_currency():
    try:
        amount = float(entry_amount.get())
        from_currency = combo_from.get()
        to_currency = combo_to.get()

        # Jika mata uang asal dan tujuan sama
        if from_currency == to_currency:
            result = amount
        else:
            rate = exchange_rates[from_currency].get(to_currency)
            if rate:
                result = amount * rate
            else:
                result_label.config(text="Nilai tukar tidak tersedia")
                return

        result_label.config(text=f'Hasil: {result:.2f} {to_currency}')
    except ValueError:
        result_label.config(text="Masukkan jumlah uang yang valid")

def reset_fields():
    entry_amount.delete(0, tk.END)
    result_label.config(text="")
    combo_from.set("")
    combo_to.set("")

root = tk.Tk()
root.title("Currency Converter")

# Input jumlah uang
tk.Label(root, text="Masukkan jumlah uang:").grid(row=0, column=0, padx=10, pady=10, sticky="w")
entry_amount = tk.Entry(root)
entry_amount.grid(row=0, column=1, padx=10, pady=10)

# Dropdown untuk mata uang asal
tk.Label(root, text="Pilih mata uang asal:").grid(row=1, column=0, padx=10, pady=10, sticky="w")
combo_from = ttk.Combobox(root, values=list(exchange_rates.keys()), state="readonly")
combo_from.grid(row=1, column=1, padx=10, pady=10)

# Dropdown untuk mata uang tujuan
tk.Label(root, text="Pilih mata uang tujuan:").grid(row=2, column=0, padx=10, pady=10, sticky="w")
combo_to = ttk.Combobox(root, values=list(exchange_rates.keys()), state="readonly")
combo_to.grid(row=2, column=1, padx=10, pady=10)

# Tombol konversi
convert_button = tk.Button(root, text="Konversi", command=convert_currency)
convert_button.grid(row=3, column=0, columnspan=2, pady=10)

# Label untuk menampilkan hasil
result_label = tk.Label(root, text="")
result_label.grid(row=4, column=0, columnspan=2, pady=10)

# Tombol reset
reset_button = tk.Button(root, text="Reset", command=reset_fields)
reset_button.grid(row=5, column=0, columnspan=2, pady=10)

root.mainloop()
```

```python
# Soal 8 Aplikasi Penjumlahan Matematika
import tkinter as tk
import random
from tkinter import messagebox

def pembelajaran():
    global num1, num2, operator, correct_answer

    # Membuat dua angka acak
    num1 = random.randint(1, 20)
    num2 = random.randint(1, 20)

    # Memilih operasi acak: penjumlahan atau pengurangan
    operator = random.choice(["+", "-"])

    if operator == "+":
        correct_answer = num1 + num2
    else:
        correct_answer = num1 - num2

    # Menampilkan soal
    label_question.config(text=f"Apa hasil dari {num1} {operator} {num2}?")

def check_answer():
    try:
        user_answer = int(entry_answer.get())
        if user_answer == correct_answer:
            messagebox.showinfo("Hasil", "Jawaban benar!")
        else:
            messagebox.showerror("Hasil", f"Jawaban salah! Jawaban yang benar adalah {correct_answer}.")
    except ValueError:
        messagebox.showerror("Error", "Masukkan jawaban dalam bentuk angka.")

    entry_answer.delete(0, tk.END)
    pembelajaran()  # Membuat soal baru

# GUI setup
root = tk.Tk()
root.title("Latihan Matematika Anak")

# Label untuk pertanyaan
label_question = tk.Label(root, text="", font=('Arial', 14))
label_question.pack(pady=10)

# Input untuk jawaban
entry_answer = tk.Entry(root, font=('Arial', 14))
entry_answer.pack(pady=10)

# Tombol untuk memeriksa jawaban
button_check = tk.Button(root, text="Cek Jawaban", command=check_answer, font=('Arial', 14))
button_check.pack(pady=10)

# Mulai dengan soal pertama
pembelajaran()

root.mainloop()
```

```python
#Soal 9 Penghitung Waktu Mundur
import tkinter as tk
from tkinter import messagebox

def start_countdown():
    try:
        countdown_time = int(entry_time.get())
        countdown(countdown_time)
    except ValueError:
        messagebox.showerror("Error", "Masukkan angka yang valid.")

def countdown(time_left):
    if time_left > 0:
        label_time.config(text=f"Waktu tersisa: {time_left} detik")
        root.after(1000, countdown, time_left - 1)
    else:
        label_time.config(text="Waktu tersisa: 0 detik")
        messagebox.showinfo("Waktu Habis", "Waktu Habis!")

# GUI setup
root = tk.Tk()
root.title("Penghitung Waktu Mundur")

# Label untuk instruksi
label_instruction = tk.Label(root, text="Masukkan waktu dalam detik:", font=('Arial', 12))
label_instruction.grid(row=0, column=0, columnspan=2, pady=10)

# Input untuk memasukkan jumlah detik
entry_time = tk.Entry(root, font=('Arial', 12))
entry_time.grid(row=1, column=0, columnspan=2, pady=5)

# Tombol untuk memulai hitungan mundur
button_start = tk.Button(root, text="Mulai", command=start_countdown, font=('Arial', 12))
button_start.grid(row=2, column=0, columnspan=2, pady=10)

# Label untuk menampilkan waktu yang tersisa
label_time = tk.Label(root, text="Waktu tersisa: 0 detik", font=('Arial', 12))
label_time.grid(row=3, column=0, columnspan=2, pady=10)

root.mainloop()
```

```python
import mysql.connector
from tkinter import *
from tkinter import messagebox

# Koneksi ke database MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="book_management"
)

cursor = conn.cursor()

# Fungsi untuk menambah buku
def add_book():
    title = title_entry.get()
    author = author_entry.get()
    year = year_entry.get()

    if title and author and year:
        query = "INSERT INTO books (title, author, year) VALUES (%s, %s, %s)"
        values = (title, author, year)
        cursor.execute(query, values)
        conn.commit()
        display_books()
        clear_entries()
        messagebox.showinfo("Success", "Buku berhasil ditambahkan!")
    else:
        messagebox.showwarning("Input Error", "Semua kolom harus diisi!")

# Fungsi untuk menghapus buku
def delete_book():
    selected_book = book_list.curselection()
    if selected_book:
        book_id = book_list.get(selected_book).split()[0]
        query = "DELETE FROM books WHERE id = %s"
        cursor.execute(query, (book_id,))
        conn.commit()
        display_books()
        messagebox.showinfo("Success", "Buku berhasil dihapus!")
    else:
        messagebox.showwarning("Selection Error", "Pilih buku yang ingin dihapus!")

# Fungsi untuk memperbarui buku
def update_book():
    selected_book = book_list.curselection()
    if selected_book:
        book_id = book_list.get(selected_book).split()[0]
        title = title_entry.get()
        author = author_entry.get()
        year = year_entry.get()

        if title and author and year:
            query = "UPDATE books SET title = %s, author = %s, year = %s WHERE id = %s"
            values = (title, author, year, book_id)
            cursor.execute(query, values)
            conn.commit()
            display_books()
            clear_entries()
            messagebox.showinfo("Success", "Buku berhasil diperbarui!")
        else:
            messagebox.showwarning("Input Error", "Semua kolom harus diisi!")
    else:
        messagebox.showwarning("Selection Error", "Pilih buku yang ingin diperbarui!")

# Fungsi untuk menampilkan semua buku
def display_books():
    book_list.delete(0, END)
    query = "SELECT * FROM books"
    cursor.execute(query)
    books = cursor.fetchall()

    for book in books:
        book_list.insert(END, f"{book[0]} - {book[1]} oleh {book[2]} ({book[3]})")

# Fungsi untuk mengosongkan input
def clear_entries():
    title_entry.delete(0, END)
    author_entry.delete(0, END)
    year_entry.delete(0, END)

# UI dengan Tkinter
root = Tk()
root.title("Pengelolaan Buku")

# Label dan Entry untuk Judul
Label(root, text="Judul").grid(row=0, column=0, padx=10, pady=10)
title_entry = Entry(root)
title_entry.grid(row=0, column=1, padx=10, pady=10)

# Label dan Entry untuk Pengarang
Label(root, text="Pengarang").grid(row=1, column=0, padx=10, pady=10)
author_entry = Entry(root)
author_entry.grid(row=1, column=1, padx=10, pady=10)

# Label dan Entry untuk Tahun Terbit
Label(root, text="Tahun Terbit").grid(row=2, column=0, padx=10, pady=10)
year_entry = Entry(root)
year_entry.grid(row=2, column=1, padx=10, pady=10)

# Tombol untuk Tambah, Hapus, Perbarui, dan Tampilkan Buku
Button(root, text="Tambah Buku", command=add_book).grid(row=3, column=0, padx=10, pady=10)
Button(root, text="Hapus Buku", command=delete_book).grid(row=3, column=1, padx=10, pady=10)
Button(root, text="Perbarui Buku", command=update_book).grid(row=3, column=2, padx=10, pady=10)

# Listbox untuk menampilkan daftar buku
book_list = Listbox(root, width=50)
book_list.grid(row=4, column=0, columnspan=3, padx=10, pady=10)

# Tombol untuk menampilkan daftar buku
Button(root, text="Tampilkan Buku", command=display_books).grid(row=5, column=0, columnspan=3, padx=10, pady=10)

# Jalankan aplikasi
root.mainloop()

# Tutup koneksi MySQL setelah aplikasi selesai
conn.close()
```
