```python
#Latihan 1 menghitung balok
import tkinter as tk

def volume_balok():
    root = tk.Tk()
    root.title('Volume Balok')

    def hitung_volume():
        panjang = float(entry_panjang.get())
        lebar = float(entry_lebar.get())
        tinggi = float(entry_tinggi.get())
        volume = panjang * lebar * tinggi
        label_hasil.config(text=f'Volume: {volume}')

    tk.Label(root, text='Panjang:').pack()
    entry_panjang = tk.Entry(root)
    entry_panjang.pack()

    tk.Label(root, text='Lebar:').pack()
    entry_lebar = tk.Entry(root)
    entry_lebar.pack()

    tk.Label(root, text='Tinggi:').pack()
    entry_tinggi = tk.Entry(root)
    entry_tinggi.pack()

    tombol_hitung = tk.Button(root, text='Hitung Volume', command=hitung_volume)
    tombol_hitung.pack()

    label_hasil = tk.Label(root, text='Volume: ')
    label_hasil.pack()

    root.mainloop()

volume_balok()
```

```python
#Latihan 2 mencari bilangan prima
import tkinter as tk

def cek_prima():
    root = tk.Tk()
    root.title('Cek Bilangan Prima')

    def is_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

    def cek_bilangan():
        angka = int(entry_angka.get())
        if is_prime(angka):
            label_hasil.config(text=f'{angka} adalah bilangan prima.')
        else:
            label_hasil.config(text=f'{angka} bukan bilangan prima.')

    tk.Label(root, text='Masukkan Angka:').pack()
    entry_angka = tk.Entry(root)
    entry_angka.pack()

    tombol_cek = tk.Button(root, text='Cek Bilangan', command=cek_bilangan)
    tombol_cek.pack()

    label_hasil = tk.Label(root, text='Hasil: ')
    label_hasil.pack()

    root.mainloop()

cek_prima()
```

```python
def latihan_9():
    root = tk.Tk()
    root.title('Latihan 9: Skala')
    def update_value(value):
        label_value.config(text=f'Nilai: {value}')
    scale = tk.Scale(root, from_=0, to=100, orient=tk.HORIZONTAL, command=update_value)
    scale.pack()
    label_value = tk.Label(root, text='Nilai: 0')
    label_value.pack()
    root.mainloop()

latihan_9()
```

```python
import tkinter as tk
from tkinter import ttk
def latihan_10():
    root = tk.Tk()
    root.title('Latihan 10: Tabel Data')
    tree = ttk.Treeview(root, columns=('Nama', 'Umur', 'Kota'), show='headings')
    tree.heading('Nama', text='Nama')
    tree.heading('Umur', text='Umur')
    tree.heading('Kota', text='Kota')
    data = [
        ('Sherly', 19, 'Jakarta'),
        ('Amelia', 20, 'Tangerang'),
        ('Putri', 21, 'Bandung')
    ]
    for row in data:
        tree.insert('', tk.END, values=row)
    tree.pack()
    root.mainloop()

latihan_10()
    
```

```python
def latihan_11():
    root = tk.Tk()
    root.title('Latihan 11: To-Do List')
    tasks = []
    def tambah_tugas():
        tugas = entry_tugas.get()
        if tugas:
            tasks.append(tugas)
            update_listbox()
            entry_tugas.delete(0, tk.END)
    def hapus_tugas():
        selected_task = listbox_tugas.curselection()
        if selected_task:
            tasks.pop(selected_task[0])
            update_listbox()
    def update_listbox():
        listbox_tugas.delete(0, tk.END)
        for task in tasks:
            listbox_tugas.insert(tk.END, task)
    entry_tugas = tk.Entry(root)
    entry_tugas.pack()
    tombol_tambah = tk.Button(root, text='Tambah Tugas', command=tambah_tugas)
    tombol_tambah.pack()
    tombol_hapus = tk.Button(root, text='Hapus Tugas', command=hapus_tugas)
    tombol_hapus.pack()
    listbox_tugas = tk.Listbox(root)
    listbox_tugas.pack()
    root.mainloop()

latihan_11()
```

```python
from tkinter import messagebox
def latihan_12():
    root = tk.Tk()
    root.title('Latihan 12: Buku Telepon')
    kontak = []
    def tambah_kontak():
        nama = entry_nama.get()
        nomor = entry_nomor.get()
        if nama and nomor:
            kontak.append((nama, nomor))
            update_listbox()
            entry_nama.delete(0, tk.END)
            entry_nomor.delete(0, tk.END)
    def hapus_kontak():
        selected_kontak = listbox_kontak.curselection()
        if selected_kontak:
            kontak.pop(selected_kontak[0])
            update_listbox()
    def update_listbox():
        listbox_kontak.delete(0, tk.END)
        for item in kontak:
            listbox_kontak.insert(tk.END, f'{item[0]} - {item[1]}')
    def edit_kontak():
        selected_kontak = listbox_kontak.curselection()
        if selected_kontak:
            nama_baru = entry_nama.get()
            nomor_baru = entry_nomor.get()
            if nama_baru and nomor_baru:
                kontak[selected_kontak[0]] = (nama_baru, nomor_baru)
                update_listbox()
                entry_nama.delete(0, tk.END)
                entry_nomor.delete(0, tk.END)
                
    tk.Label(root, text="Nama:").pack()
    entry_nama = tk.Entry(root)
    entry_nama.pack()

    tk.Label(root, text="Nomor Telepon:").pack()
    entry_nomor = tk.Entry(root)
    entry_nomor.pack()

    frame_tombol = tk.Frame(root)
    frame_tombol.pack()

    tombol_tambah = tk.Button(frame_tombol, text="Tambah", command=tambah_kontak)
    tombol_tambah.grid(row=0, column=0)

    tombol_edit = tk.Button(frame_tombol, text="Edit", command=edit_kontak)
    tombol_edit.grid(row=0, column=1)

    tombol_hapus = tk.Button(frame_tombol, text="Hapus", command=hapus_kontak)
    tombol_hapus.grid(row=0, column=2)

    listbox_kontak = tk.Listbox(root)
    listbox_kontak.pack()

    root.mainloop()

latihan_12()
```
